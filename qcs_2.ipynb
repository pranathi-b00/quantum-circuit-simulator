```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
```

```python
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
qc=QuantumCircuit(1, 1)
qc.h(0)
qc.measure(0, 0)
qc.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
compiled_circuit=transpile(qc, backend)
job = backend.run(compiled_circuit, shots=1024)
result = job.result()
counts = result.get_counts()
plot_histogram(counts)
plt.show()
```

```python
qc = QuantumCircuit(1, 1)

qc.x(0)              # applying X gate
qc.measure(0, 0)

compiled_qc = transpile(qc, backend)
job = backend.run(compiled_qc, shots=1024)
result = job.result()

counts = result.get_counts()
print(counts)

plot_histogram(counts)
plt.show()
```

```python
qc = QuantumCircuit(1, 1)

qc.h(0)      # creating superposition
qc.z(0)      # applying phase flip or Z gate
qc.h(0)      # bringing back to computational basis
qc.measure(0, 0)

compiled_qc = transpile(qc, backend)
job = backend.run(compiled_qc, shots=1024)
result = job.result()

counts = result.get_counts()
print(counts)

plot_histogram(counts)
plt.show()
```

```python
qc = QuantumCircuit(2, 2)

qc.h(0)          # now put qubit 0 in superposition
qc.cx(0, 1)      # entangling qubit 1 with qubit 0
qc.measure([0, 1], [0, 1])

compiled_qc = transpile(qc, backend)
job = backend.run(compiled_qc, shots=1024)
result = job.result()

counts = result.get_counts()
print(counts)

plot_histogram(counts)
plt.show()
```
